```python
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the dataset from a CSV file
file_name = 'simulated_energy_data.csv'
data = pd.read_csv(file_name)
```

```python
# display the DataFrame's information and first few rows
print(data.info())
print(data.head())
```

```python
# display a statistical summary of the DataFrame
print(data.describe())
```

Note the set ranges
energy consumtpion: 5-15
temp: 20 - 80
humidity %: 30 to 90

```python
# histograms to explore data distributions
data.hist(bins=30, figsize=(10, 7))
plt.tight_layout()
plt.show()
```

```python
# line plot for energy consumption over time
plt.figure(figsize=(10, 5))
plt.plot(data['Timestamp'], data['Energy_Consumption_kWh'], label='Energy Consumption')
plt.xlabel('Time')
plt.ylabel('Energy Consumption (kWh)')
plt.title('Energy Consumption Over Time')
plt.legend()
plt.show()
```
